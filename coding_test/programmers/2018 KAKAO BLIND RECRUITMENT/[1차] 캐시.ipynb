```txt
문제 설명
캐시
지도개발팀에서 근무하는 제이지는 지도에서 도시 이름을 검색하면 해당 도시와 관련된 맛집 게시물들을 데이터베이스에서 읽어 보여주는 서비스를 개발하고 있다.
이 프로그램의 테스팅 업무를 담당하고 있는 어피치는 서비스를 오픈하기 전 각 로직에 대한 성능 측정을 수행하였는데, 제이지가 작성한 부분 중 데이터베이스에서 게시물을 가져오는 부분의 실행시간이 너무 오래 걸린다는 것을 알게 되었다.
어피치는 제이지에게 해당 로직을 개선하라고 닦달하기 시작하였고, 제이지는 DB 캐시를 적용하여 성능 개선을 시도하고 있지만 캐시 크기를 얼마로 해야 효율적인지 몰라 난감한 상황이다.

어피치에게 시달리는 제이지를 도와, DB 캐시를 적용할 때 캐시 크기에 따른 실행시간 측정 프로그램을 작성하시오.

입력 형식
- 캐시 크기(cacheSize)와 도시이름 배열(cities)을 입력받는다.
- cacheSize는 정수이며, 범위는 0 ≦ cacheSize ≦ 30 이다.
- cities는 도시 이름으로 이뤄진 문자열 배열로, 최대 도시 수는 100,000개이다.
- 각 도시 이름은 공백, 숫자, 특수문자 등이 없는 영문자로 구성되며, 대소문자 구분을 하지 않는다. 도시 이름은 최대 20자로 이루어져 있다.

출력 형식
- 입력된 도시이름 배열을 순서대로 처리할 때, "총 실행시간"을 출력한다.

조건
- 캐시 교체 알고리즘은 LRU(Least Recently Used)를 사용한다.
- cache hit일 경우 실행시간은 1이다.
- cache miss일 경우 실행시간은 5이다.

입출력 예제
캐시크기(cacheSize)	도시이름(cities)	실행시간
3	["Jeju", "Pangyo", "Seoul", "NewYork", "LA", "Jeju", "Pangyo", "Seoul", "NewYork", "LA"]	50
3	["Jeju", "Pangyo", "Seoul", "Jeju", "Pangyo", "Seoul", "Jeju", "Pangyo", "Seoul"]	21
2	["Jeju", "Pangyo", "Seoul", "NewYork", "LA", "SanFrancisco", "Seoul", "Rome", "Paris", "Jeju", "NewYork", "Rome"]	60
5	["Jeju", "Pangyo", "Seoul", "NewYork", "LA", "SanFrancisco", "Seoul", "Rome", "Paris", "Jeju", "NewYork", "Rome"]	52
2	["Jeju", "Pangyo", "NewYork", "newyork"]	16
0	["Jeju", "Pangyo", "Seoul", "NewYork", "LA"]	25
```

````python
from collections import OrderedDict

# 나는 deque + dict로 캐시 힛을 구현했는데, 문제는 이렇게 하면 캐시 힛을 하는 경우 deque에서 최근에 들어온 단어를 다시 최신 상태로 만들어주는 과정에서 deque 길이 만큼 O(n)이 발생한다. 
# 그래서 정석 방법으로 OrderedDict를 사용하는 방법이 있다고 한다. 순서가 있는 dict 이다. 
# 구현 방법은 동일하다. 
# 1. 캐시 힛이면 카운트 +1 해주고, 해당 단어를 최신 상태로 만들어준다.
# 최신 상태로 만들려면 OrderedDict.move_to_end(last = False)로 하면 맨 뒤로 이동한다(last = True이면 맨 앞으로 이동).
# 2. 캐시 힛이 아니면 
#   1) 캐시에 여유가 있다면 그냥 추가
#   2) 없다면 가장 과거 캐시 지우고 추가
#   위 과정 후 카운트 +5

def solution(cacheSize, cities):
    if cacheSize == 0:
        return len(cities) * 5
    answer = 0
    dicCache = OrderedDict()
    for i in cities : 
        i = i.lower()
        
        # 캐시에 있을 때
        if i in dicCache : 
            answer += 1
            dicCache.move_to_end(i, last = True)       # 맨 뒤로 이동
            continue
        
        # 캐시에 없을 때
        if len(dicCache) < cacheSize : 
            dicCache[i] = True
        else : 
            if dicCache : 
                dicCache.popitem(last = False)       # 맨 앞 제거
            dicCache[i] = True
        answer += 5
    
    return answer


# 답 50
cacheSize = 3
cities = ["Jeju", "Pangyo", "Seoul", "NewYork", "LA", "Jeju", "Pangyo", "Seoul", "NewYork", "LA"]
result = solution(cacheSize, cities)
print(result)

# 답 21
cacheSize = 3
cities = ["Jeju", "Pangyo", "Seoul", "Jeju", "Pangyo", "Seoul", "Jeju", "Pangyo", "Seoul"]
result = solution(cacheSize, cities)
print(result)

# 답 60
cacheSize = 2
cities = ["Jeju", "Pangyo", "Seoul", "NewYork", "LA", "SanFrancisco", "Seoul", "Rome", "Paris", "Jeju", "NewYork", "Rome"]
result = solution(cacheSize, cities)
print(result)

# 답 52
cacheSize = 5
cities = ["Jeju", "Pangyo", "Seoul", "NewYork", "LA", "SanFrancisco", "Seoul", "Rome", "Paris", "Jeju", "NewYork", "Rome"]
result = solution(cacheSize, cities)
print(result)

# 답 16
cacheSize = 2
cities = ["Jeju", "Pangyo", "NewYork", "newyork"]
result = solution(cacheSize, cities)
print(result)

# 답 25
cacheSize = 0
cities = ["Jeju", "Pangyo", "Seoul", "NewYork", "LA"]
result = solution(cacheSize, cities)
print(result)
````

````python
from collections import deque

# '캐시 교체 알고리즘은 LRU(Least Recently Used)를 사용한다.' 라는 조건은 캐시에 저장하는 사이즈가 5일 떄, 5만큼 차고나서 다음 게 들어오면 가장 과거의 것을 지우고 최근 걸 추가하는 방식이다.
# 즉, deque를 써서 popleft() 하면 된다. 하지만 캐시에 있는지 검색하려면 O(n)이 걸리기 때문에 deque로 구현하는 건 무리가 있다.
# dict로 key로 만들어서 검색하는 것이 필요하다. 검색은 이걸로 하고, 순서는 deque로 결정한다.

# 구현 방법
# 1. 들어온 순서를 기록할 dq, 문자 검색용도인 dicCache 를 만든다.
# 2. 문자가 들어오면 검사
#   1) s in dicCache : 실행시간 +1. 도시가 최근에 검색되었으므로 캐시에 해당 도시 최근으로 업데이트하기 위해 dq.remove(도시) -> dq.append(도시)
#   2) len(dq) <= cacheSize 이면 그냥 넣고, 실행시간 5 카운트
#      else : popleft() 한 걸 dicCache에서 제거, 새로 들어온 걸 append(). 실행시간 5 카운트


def solution(cacheSize, cities):
    answer = 0
    dq = deque([])
    dicCache = {}
    if cacheSize == 0 : 
        answer = len(cities) * 5
        return answer
    
    for i in cities : 
        #print(dq, dicCache)
        i = i.lower()
        # cache에 있을 떄
        if i in dicCache : 
            answer += 1
            # LRU 갱신: 해당 도시를 가장 최근으로 이동
            dq.remove(i)
            dq.append(i)
            continue
        
        # cache에 없을 때
        if len(dicCache) < cacheSize : 
            dq.append(i)
            dicCache[i] = True
        else : 
            if dq : 
                sPast = dq.popleft()
                del dicCache[sPast]
            dq.append(i)
            dicCache[i] = True
        answer += 5
    
    return answer

# 답 50
cacheSize = 3
cities = ["Jeju", "Pangyo", "Seoul", "NewYork", "LA", "Jeju", "Pangyo", "Seoul", "NewYork", "LA"]
result = solution(cacheSize, cities)
print(result)

# 답 21
cacheSize = 3
cities = ["Jeju", "Pangyo", "Seoul", "Jeju", "Pangyo", "Seoul", "Jeju", "Pangyo", "Seoul"]
result = solution(cacheSize, cities)
print(result)

# 답 60
cacheSize = 2
cities = ["Jeju", "Pangyo", "Seoul", "NewYork", "LA", "SanFrancisco", "Seoul", "Rome", "Paris", "Jeju", "NewYork", "Rome"]
result = solution(cacheSize, cities)
print(result)

# 답 52
cacheSize = 5
cities = ["Jeju", "Pangyo", "Seoul", "NewYork", "LA", "SanFrancisco", "Seoul", "Rome", "Paris", "Jeju", "NewYork", "Rome"]
result = solution(cacheSize, cities)
print(result)

# 답 16
cacheSize = 2
cities = ["Jeju", "Pangyo", "NewYork", "newyork"]
result = solution(cacheSize, cities)
print(result)

# 답 25
cacheSize = 0
cities = ["Jeju", "Pangyo", "Seoul", "NewYork", "LA"]
result = solution(cacheSize, cities)
print(result)
````
