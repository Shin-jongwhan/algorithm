```txt
문제 설명
짝지어 제거하기는, 알파벳 소문자로 이루어진 문자열을 가지고 시작합니다. 먼저 문자열에서 같은 알파벳이 2개 붙어 있는 짝을 찾습니다. 그다음, 그 둘을 제거한 뒤, 앞뒤로 문자열을 이어 붙입니다. 이 과정을 반복해서 문자열을 모두 제거한다면 짝지어 제거하기가 종료됩니다. 문자열 S가 주어졌을 때, 짝지어 제거하기를 성공적으로 수행할 수 있는지 반환하는 함수를 완성해 주세요. 성공적으로 수행할 수 있으면 1을, 아닐 경우 0을 리턴해주면 됩니다.

예를 들어, 문자열 S = baabaa 라면

b aa baa → bb aa → aa →

의 순서로 문자열을 모두 제거할 수 있으므로 1을 반환합니다.

제한사항
문자열의 길이 : 1,000,000이하의 자연수
문자열은 모두 소문자로 이루어져 있습니다.
입출력 예
s	result
baabaa	1
cdcd	0
입출력 예 설명
입출력 예 #1
위의 예시와 같습니다.
입출력 예 #2
문자열이 남아있지만 짝지어 제거할 수 있는 문자열이 더 이상 존재하지 않기 때문에 0을 반환합니다.
```


````python
# stack을 활용
# 만약 stack이 비어있으면 append(i)
# 만약 비어있지 않고 stack[-1] == i 라면 stack.pop()
# 만약 비어있지 않고 stack[-1] != i 라면 stack.append(i)

# 시간복잡도 : O(n)

def solution(s):
    if len(s) <= 1 : 
        return 0
    answer = 0

    lsStack = []
    for i in s : 
        if lsStack and lsStack[-1] == i : 
            lsStack.pop()
        else : 
            lsStack.append(i)

    if lsStack == [] : 
        answer = 1

    return answer

s = 'baabaa'
result = solution(s)
print(result)

s = 'cdcd'
result = solution(s)
print(result)
````

````python
# s[i] == s[i + 1]이면 제거
# 제거 후 i -= 1 (i > 0 인 경우만. i == 0 이면 변화 없음)
# 만약 같지 않으면 i += 1 (i < len(s) - 2 일 경우만. 아니면 break)
# 그런데 문자열을 제거하는 방식은 새롭게 문자열을 만드는 방식이라 효율성이 안 좋음.

# 시간복잡도 : 최적 O(n), 최악 O(n^2)
# n = s

# 검증 통과
# 효율성 테스트 실패
def solution(s):
    if len(s) <= 1 : 
        return 0
    answer = 0

    i = 0
    while True : 
        if len(s) == 0 or i >= len(s) - 1 : 
            break
        if s[i] == s[i + 1] : 
            s = s[:i] + s[i + 2:]
            if i > 0 : 
                i -= 1
        else :
            if i <= len(s) - 2 : 
                i += 1
    
    if s == '' : 
        answer = 1

    return answer


s = 'baabaa'
result = solution(s)
print(result)

s = 'cdcd'
result = solution(s)
print(result)
````
