문제 설명
어떤 게임에는 붕대 감기라는 기술이 있습니다.

붕대 감기는 t초 동안 붕대를 감으면서 1초마다 x만큼의 체력을 회복합니다. t초 연속으로 붕대를 감는 데 성공한다면 y만큼의 체력을 추가로 회복합니다. 게임 캐릭터에는 최대 체력이 존재해 현재 체력이 최대 체력보다 커지는 것은 불가능합니다.

기술을 쓰는 도중 몬스터에게 공격을 당하면 기술이 취소되고, 공격을 당하는 순간에는 체력을 회복할 수 없습니다. 몬스터에게 공격당해 기술이 취소당하거나 기술이 끝나면 그 즉시 붕대 감기를 다시 사용하며, 연속 성공 시간이 0으로 초기화됩니다.

몬스터의 공격을 받으면 정해진 피해량만큼 현재 체력이 줄어듭니다. 이때, 현재 체력이 0 이하가 되면 캐릭터가 죽으며 더 이상 체력을 회복할 수 없습니다.

당신은 붕대감기 기술의 정보, 캐릭터가 가진 최대 체력과 몬스터의 공격 패턴이 주어질 때 캐릭터가 끝까지 생존할 수 있는지 궁금합니다.

1 ≤ 시전 시간 = t ≤ 50
1 ≤ 초당 회복량 = x ≤ 100
1 ≤ 추가 회복량 = y ≤ 100
1 ≤ health ≤ 1,000
1 ≤ attacks의 길이 ≤ 100
attacks[i]는 [공격 시간, 피해량] 형태의 길이가 2인 정수 배열입니다.
attacks는 공격 시간을 기준으로 오름차순 정렬된 상태입니다.
attacks의 공격 시간은 모두 다릅니다.
1 ≤ 공격 시간 ≤ 1,000
1 ≤ 피해량 ≤ 100
입출력 예

몬스터의 마지막 공격은 3초에 이루어집니다. 0초부터 3초까지 캐릭터의 상태는 아래 표와 같습니다.

몬스터의 마지막 공격 직후 캐릭터의 체력은 3입니다. 따라서 3을 return 해야 합니다.

````python
# 붕대 성공 -> 체력 회복
# 붕대 t초 연속 성공 -> 추가 체력 회복 + 붕대 체력 회복만큼 회복, 붕대 연속 성공 초기화
# 만약 현재 체력 + 체력 회복 > 최대 체력이면 최대 체력으로 유지
# 공격을 받으면 -> 회복량 없음, 공격 받은 만큼 체력 마이너스, 붕대 연속 성공 초기화
# 캐릭터 시작 체력에 대한 설명이 없고, 예제에서 최대 체력으로 시작하였기 때문에 동일하게  구현

# bandage는 [시전 시간, 초당 회복량, 추가 회복량]
# attacks[i]는 [공격 시간, 피해량]

# return
# 모든 공격이 끝난 직후 남은 체력을 return. 이후의 체력 회복은 없음.
# 만약 몬스터의 공격을 받고 캐릭터의 체력이 0 이하가 되어 죽는다면 -1을 return

# attacks는 같은 시간 2회 이상 공격이 없기 때문에 이점은 고려 X
# 시간이 오름차순으로 되어 있기 때문에 마지막 원소를 전체 시간 nSec으로 고려하면 됨
# nSec 으로 for loop 돌면서 로직 구현

# 시간복잡도 : O(n)
# nSec = n

def solution(bandage, health, attacks):
    answer = 0
    
    health_current = health
    attacks = list(reversed(attacks))     # 시간 순으로 pop() 하기 위함
    nConinue_heal = 0      # 연속 회복 횟수
    
    nSec = attacks[0][0]
    for i in range(1, nSec + 1) : 
        # 공격을 받으면 회복할 수 없기 때문에 공격 먼저 고려
        if i == attacks[-1][0] : 
            health_current -= attacks[-1][1]
            attacks.pop()
            nConinue_heal = 0       # 연속 회복 초기화
            
            # 죽음
            if health_current <= 0 : 
                answer = -1
                return answer
            continue
        
        # 회복
        health_current += bandage[1]
        nConinue_heal += 1
        # 추가 회복
        if nConinue_heal == bandage[0] : 
            health_current += bandage[2]
            nConinue_heal = 0       # 연속 회복 초기화
        # 최대 체력
        if health_current > health : 
            health_current = health
            
    answer = health_current
    
    return answer


bandage = [5, 1, 5]
health = 30
attacks = [[2, 10], [9, 15], [10, 5], [11, 5]]
# result 5
result = solution(bandage, health, attacks)
print(result)

bandage = [3, 2, 7]
health = 20
attacks = [[1, 15], [5, 16], [8, 6]]
# result -1
result = solution(bandage, health, attacks)
print(result)

bandage = [4, 2, 7]
health = 20
attacks = [[1, 15], [5, 16], [8, 6]]
# result -1
result = solution(bandage, health, attacks)
print(result)

bandage = [1, 1, 1]
health = 5
attacks = [[1, 2], [3, 2]]
# result 3
result = solution(bandage, health, attacks)
print(result)
````
