### 방법

```txt
같은 i에 또는 같은 j에 Q이 1개 이상일 수 없다. 즉, 이미 i / j 자리를 차지하고 있으면 다른 자리에 가야 한다.
그리고 i +- n, j +- n에도 올 수 없다. 대각선 자리이다.

여기서 대각선 좌표에 대한 수학적 규칙을 알고 있으면 풀기 쉽다.
여기서 대각선은 2가지 방향이 있다. '/' 방향, '\' 방향이 있다.
'/' 방향 : 같은 대각선 방향이라면 row + col이 성립한다.
'\' 방향 : 같은 대각선 방향이라면 row - col이 성립한다.

대각선 방향의 개수
'/' 방향
최소 [0, 0] -> row + col = 0
최대 [n - 1, n - 1] -> row + col = 2n - 2
0 부터 2n - 2까지 총 2n - 1개가 있다.
'\' 방향
최소 [0, n - 1] -> row - col = -n + 1
최대 [n - 1, 0] -> row - col = n - 1
-n + 1부터 n - 1까지 총 2n - 1개가 있다. 

풀이 방법
1. j열이 갈 수 있는 자리인지 검사하는 리스트 하나 생성. [False] * n 으로 초기화 -> col idx 그대로 접근 가능
2. 대각선 '/', '\' 방향에 대해 갈 수 있는 자리인지 검사한다. [False] * (2n - 1)로 초기화
  1) '/' 방향 대각선 : row + col로 접근 가능
  2) '\' 방향 대각선 : row - col + (n - 1)로 접근 가능. (n - 1)을 더해주는 해주는 최솟값이 row - col의 최솟값이 -n + 1이기 떄문에 리스트에 접근할 떄 인덱스 에러가 나기 때문
3. 행 row를 고정하여 queen이 들어갈 수 있는지 검사하는 함수 n_queen(row)를 만든다.
접근한 row == n 이라면 모두 다 배치하였다는 뜻이므로 answer += 1 하고 return
col을 하나씩 순회한다. 
  1) 열, 대각선에 접근 가능한지 체크(False) -> 접근 불가능하면 continue
  2) 접근 가능하면 열, 대각선 접근 가능을 True로 만든다(접근 불가).
  3) 그 다음 행인 row + 1에 배치하기 위해 n_queen(row) 재귀함수 호출
  4) 순회가 끝나면 다시 돌아온 상태이므로 열, 대각선 접근 가능을 False로 다시 변경.
```


````python
def solution(n):
    answer = 0
    nUsed_col = [False] * n
    nUsed_diag_1 = [False] * (2 * n - 1)        # '/' 방향
    nUsed_diag_2 = [False] * (2 * n - 1)        # '\' 방향
    
    def n_queen(row) : 
        nonlocal answer     # 바깥에 있는 answer 값을 바꿀 수 있도록 처리
        # 모든 row에 배치한 상태이므로 +1 하고 리턴
        if row == n : 
            answer += 1
            return
        
        for col in range(0, n) : 
            nDiag_1 = row + col
            nDiag_2 = row - col + (n - 1)
            
            if nUsed_col[col] == True : 
                continue
            elif nUsed_diag_1[nDiag_1] == True : 
                continue
            elif nUsed_diag_2[nDiag_2] == True : 
                continue
            
            # 접근 불가 처리
            nUsed_col[col] = True
            nUsed_diag_1[nDiag_1] = True
            nUsed_diag_2[nDiag_2] = True
            
            # 다음 row로 진행
            n_queen(row + 1)
            
            # 순회하고 다시 돌아왔으므로 접근 가능 처리
            nUsed_col[col] = False
            nUsed_diag_1[nDiag_1] = False
            nUsed_diag_2[nDiag_2] = False
            
    # 0번째 행부터 검사
    n_queen(0)
    
    return answer


n = 4
result = solution(n)
print(result)
````
