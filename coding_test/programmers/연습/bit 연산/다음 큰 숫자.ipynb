```txt
문제 설명
자연수 n이 주어졌을 때, n의 다음 큰 숫자는 다음과 같이 정의 합니다.

조건 1. n의 다음 큰 숫자는 n보다 큰 자연수 입니다.
조건 2. n의 다음 큰 숫자와 n은 2진수로 변환했을 때 1의 갯수가 같습니다.
조건 3. n의 다음 큰 숫자는 조건 1, 2를 만족하는 수 중 가장 작은 수 입니다.
예를 들어서 78(1001110)의 다음 큰 숫자는 83(1010011)입니다.

자연수 n이 매개변수로 주어질 때, n의 다음 큰 숫자를 return 하는 solution 함수를 완성해주세요.

제한 사항
n은 1,000,000 이하의 자연수 입니다.
입출력 예
n	result
78	83
15	23
입출력 예 설명
입출력 예#1
문제 예시와 같습니다.
입출력 예#2
15(1111)의 다음 큰 숫자는 23(10111)입니다.
```

````python
# 1의 개수는 같으므로 이진수 가장 큰 자릿수의 다음 자릿수에서부터 0을 만난 후, 다음 1을 만날 때까지 진행하다가 바로 이전 것을 1로 바꾸고, 만난 1을 0으로 바꾸기. 그 다음 자릿수들의 1개수 만큼을 가장 낮을 자릿수로 채우면 그 다음 가장 큰 수가 됨
# -> 이렇게 하면 큰 수부터 검색하므로 그 하위에서 바꿀 수 있으면 에러케이스가 되서 안 됨
# 아이디어는 동일한데, 오른쪽(작은 수)부터 01을 찾아서 10으로 바꿔줘야 함. 그 하위는 1 개수만큼 모두 오른쪽으로 밀어버리는 건 똑같음
# 111111과 같이 모두 1이면 0이 없으므로 가장 큰 자릿수에 1을 추가하고 그 다음 자릿수를 0으로 바꾸기

# s = bin(10)[2:]으로 하면 이진수 문자열을 출력할 수 있음.
# 다시 10진수로 변환하려면 int(s, 2)

# 1001111

# 참고
# 여기서 숫자 0은 고려 대상이 아님. 위 조건이 성립할 수 없으므로 1 이상의 자연수로 봐야 한다.
# 가장 큰 자릿수는 이진수로 치환하면 자릿수가 고정이 아닌 이상 항상 1이다. 따라서 1번째 자릿수부터 고려한다.

def solution(nNum):
    if nNum == 1:
        return 2

    lsBin = list('0' + bin(nNum)[2:])
    nLen = len(lsBin)

    nIdx_swap = -1
    for nIdx in range(nLen - 1, 0, -1):
        if lsBin[nIdx - 1] == '0' and lsBin[nIdx] == '1':
            nIdx_swap = nIdx - 1
            break

    lsBin[nIdx_swap] = '1'
    lsBin[nIdx_swap + 1] = '0'

    lsSuffix = lsBin[nIdx_swap + 2:]
    nOne_num = 0
    for sCh in lsSuffix:
        if sCh == '1':
            nOne_num += 1

    nZero_num = len(lsSuffix) - nOne_num
    lsBin = lsBin[:nIdx_swap + 2] + (['0'] * nZero_num) + (['1'] * nOne_num)

    if lsBin[0] == '0':
        lsBin = lsBin[1:]

    return int("".join(lsBin), 2)


# 답 83
n = 78
result = solution(n)
print(result)

# 답 23
n = 15
result = solution(n)
print(result)
````

````python
# 참고 : 비트연산으로도 풀 수 있다. 코드는 간단한데, 내가 작성한 코드의 방법을 비트 연산 형식으로 그대로 구현을 할 수 있음. 다만 익숙하지 않아서 사용하는 건 좀 어려울 듯.

def solution(nNum):
    # nNum은 1 이상의 자연수라고 가정하지만, 방어적으로 0이면 0 반환
    if nNum == 0:
        return 0

    # 1) 가장 낮은 자리(LSB)에 있는 '1' 비트 하나만 남긴 값 추출
    #    - -nNum은 부호를 전부 뒤집고 1을 더한다. 
    #    - 예: 1011000 -> 0001000
    #    - n & -n (2의 보수 성질)로 가장 오른쪽 1만 분리 가능
    nSmallest_one = nNum & -nNum

    # 2) 그 '1'을 더해서, 오른쪽에서 가장 가까운 "01 -> 10" 변화(캐리 포함)를 만든다.
    #    - 예: 1011000 + 0001000 = 1100000
    #    - 이 단계에서 "다음 큰 수"의 상위 비트 형태(골격)가 결정된다.
    nRipple = nNum + nSmallest_one

    # 3) nNum에서 nRipple로 바뀌는 동안 뒤집힌 비트 구간을 XOR로 표시한다.
    #    - XOR는 서로 다른 비트만 1로 만든다.
    #    - 즉, 캐리로 인해 변한 구간(정리된 부분)을 마스크로 얻는다.
    #    - 예: 1011000 ^ 1100000 = 0111000
    nOnes = nNum ^ nRipple

    # 4) 바뀐 구간 안에 있던 '1'들을 "맨 아래 비트(LSB) 쪽"으로 몰아 넣기 위한 패턴을 만든다.
    #    - "01 -> 10"으로 바뀐 두 비트는 이미 자리 확정이므로 제외하기 위해 >> 2 한다.
    #    - 그 다음 nSmallest_one(2^k 형태)로 나누면 결과적으로 k칸 오른쪽으로 이동한 효과가 난다.
    #    - 이렇게 하면 남은 '1'들이 가장 오른쪽으로 정렬된 형태가 된다.
    #    - 예: nOnes=0111000
    #         (0111000 >> 2) = 0001110
    #         0001110 // 0001000 = 0000001
    nOnes = (nOnes >> 2) // nSmallest_one

    # 5) nRipple(상위 골격) + nOnes(아래로 몰린 1들)을 OR로 합치면 정답
    #    - OR는 서로 다른 위치의 1들을 합쳐 최종 비트 패턴을 만든다.
    #    - 결과는 (1의 개수 동일) + (nNum보다 큼) + (그 중 최소) 를 만족한다.
    nAnswer = nRipple | nOnes
    return nAnswer
````
