### 문제
#### https://school.programmers.co.kr/learn/courses/30/lessons/42583

```txt
문제 설명
트럭 여러 대가 강을 가로지르는 일차선 다리를 정해진 순으로 건너려 합니다. 모든 트럭이 다리를 건너려면 최소 몇 초가 걸리는지 알아내야 합니다. 다리에는 트럭이 최대 bridge_length대 올라갈 수 있으며, 다리는 weight 이하까지의 무게를 견딜 수 있습니다. 단, 다리에 완전히 오르지 않은 트럭의 무게는 무시합니다.

예를 들어, 트럭 2대가 올라갈 수 있고 무게를 10kg까지 견디는 다리가 있습니다. 무게가 [7, 4, 5, 6]kg인 트럭이 순서대로 최단 시간 안에 다리를 건너려면 다음과 같이 건너야 합니다.

[시뮬레이션 진행 표]
| 경과 시간  |  다리를 지난 트럭  |   다리를 건너는 트럭   | 대기 트럭 |
|----------:|------------------|--------------------|-----------|
| 0         | []               | []                 | [7,4,5,6] |
| 1~2       | []               | [7]                | [4,5,6]   |
| 3         | [7]              | [4]                | [5,6]     |
| 4         | [7]              | [4,5]              | [6]       |
| 5         | [7,4]            | [5]                | [6]       |
| 6~7       | [7,4,5]          | [6]                | []        |
| 8         | [7,4,5,6]        | []                 | []        |

따라서, 모든 트럭이 다리를 지나려면 최소 8초가 걸립니다.

solution 함수의 매개변수로 다리에 올라갈 수 있는 트럭 수 bridge_length, 다리가 견딜 수 있는 무게 weight, 트럭 별 무게 truck_weights가 주어집니다. 이때 모든 트럭이 다리를 건너려면 최소 몇 초가 걸리는지 return 하도록 solution 함수를 완성하세요.

제한 조건
- bridge_length는 1 이상 10,000 이하입니다.
- weight는 1 이상 10,000 이하입니다.
- truck_weights의 길이는 1 이상 10,000 이하입니다.
- 모든 트럭의 무게는 1 이상 weight 이하입니다.

입출력 예
| bridge_length | weight | truck_weights                            | return |
|--------------:|-------:|------------------------------------------|-------:|
| 2             | 10     | [7,4,5,6]                                | 8      |
| 100           | 100    | [10]                                     | 101    |
| 100           | 100    | [10,10,10,10,10,10,10,10,10,10]          | 110    |
```

### 풀이
```txt
대기 트럭의 순서대로 다리로 가야 한다. -> 대기 트럭을 deque로 만들고 popleft()
같은 시간에 트럭이 빠지고 들어가는 건 가능하다. 
bridge_length 만큼 이동시간이 있다.

방법
1. 데이터 생성
  1) bridge_length 만큼의 길이로 0으로 초기화된 deque를 만든다. 이게 '다리를 건너는 트럭'에 대한 변수가 된다. 이를 dqLoaded 라고 하겠다.
  2) bridge에 올라간 weight를 세는 변수 nLoaded_weight = 0으로 초기화
  3) truck_weights를 deque로 만든다. 이게 대기 트럭 리스트가 된다.
  4) 다리를 지난 트럭 리스트 생성 (없어도 문제는 풀 수 있음)
2. while 문으로 대기 트럭이 있다면
  대기 트럭 popleft()
  만약 대기 트럭 nLoaded_weight + weight < nLoaded_weight 라면 -> popleft() 하고 1 이상이라면 다리를 지난 트럭에 추가 -> nLoaded_weight - (다리를 지난 트럭 weight), 대기 트럭을 append()
```

````python
from collections import deque


def solution(bridge_length, weight, truck_weights):
    answer = 0
    dqLoaded = deque([0] * bridge_length)
    nLoaded_weight = 0
    dqTruck_weight = deque(truck_weights)
    lsTruck_passed = []
    
    while dqTruck_weight : 
        nTruck_weight = dqTruck_weight.popleft()
        nTruck_weight_passed = dqLoaded.popleft()
        nLoaded_weight -= nTruck_weight_passed
        if nLoaded_weight + nTruck_weight <= weight : 
            dqLoaded.append(nTruck_weight)
            nLoaded_weight += nTruck_weight
            if nTruck_weight_passed != 0 : 
                lsTruck_passed.append(nTruck_weight_passed)
        else : 
            dqLoaded.append(0)
            dqTruck_weight.appendleft(nTruck_weight)
            
        answer += 1
    
    # 마지막 트럭 지난 시간
    answer += bridge_length
    
    return answer


# 답 8
bridge_length = 2
weight = 10
truck_weights = [7,4,5,6]
result = solution(bridge_length, weight, truck_weights)
print(result)

# 답 101
bridge_length = 100
weight = 100
truck_weights = [10]
result = solution(bridge_length, weight, truck_weights)
print(result)

# 답 110
bridge_length = 100
weight = 100
truck_weights = [10,10,10,10,10,10,10,10,10,10]
result = solution(bridge_length, weight, truck_weights)
print(result)
````
