```python
# ë…¼ë¬¸ ë¦¬ìŠ¤íŠ¸ë¥¼ ì¸ìš© ìˆ˜ë¡œ ì—­ìˆœ(ë‚´ë¦¼ì°¨ìˆœ)ìœ¼ë¡œ ì •ë ¬
# for loop -> idx + 1ì´ ë…¼ë¬¸ ìˆ˜
# ì¸ìš© ìˆ˜ ì—­ìˆœìœ¼ë¡œ í•˜ë¯€ë¡œ idxëŠ” 1ì”© ì»¤ì§€ê³ , ì¸ìš© ìˆ˜ëŠ” ê³„ì† ìž‘ì•„ì§
# idx ë²ˆì§¸ ë…¼ë¬¸ ì¸ìš© ìˆ˜ >= idx + 1 ì´ë©´ H-index = idx + 1
# idx ë²ˆì§¸ ë…¼ë¬¸ ì¸ìš© ìˆ˜ê°€ idx + 1 ë³´ë‹¤ ì ì€ ë…¼ë¬¸ì„ ë§Œë‚˜ë©´ H-indexê°€ ë˜ì§€ ëª»í•˜ë¯€ë¡œ break

# ì‹œê°„ë³µìž¡ë„ : O(n^2 * logn)
# ë…¼ë¬¸ ìˆ˜ n
# sort O(n * logn)

def solution(citations):
    lsSort = sorted(citations, reverse=True)

    nH = 0
    for nIdx in range(0, len(lsSort)) :
        nCite = lsSort[nIdx]
        print(nIdx, nCite)
        nCandidate = nIdx + 1
        # ìƒìœ„ nCandidateíŽ¸ ì¤‘ì—ì„œ, ê°€ìž¥ ë‚®ì€ ì¸ìš© ìˆ˜ê°€ nCandidate ì´ìƒ
        if nCite >= nCandidate:
            nH = nCandidate
        else:
            break

    return nH


# ë‹µ 3
citations = [3, 0, 6, 1, 5]
result = solution(citations)
print("result", result)

# ë‹µ 2
citations = [1, 2, 3, 0, 10000]
result = solution(citations)
print("result", result)

# ë‹µ 3
citations = [1, 2, 3, 4, 5]
result = solution(citations)
print("result", result)

# ë‹µ 3
citations = [0, 0, 0, 6, 6, 6]
result = solution(citations)
print("result", result)
```
