### 문제
#### https://school.programmers.co.kr/learn/courses/30/lessons/42883

```txt
문제 설명
어떤 숫자에서 k개의 수를 제거했을 때 얻을 수 있는 가장 큰 숫자를 구하려 합니다.

예를 들어, 숫자 1924에서 수 두 개를 제거하면 [19, 12, 14, 92, 94, 24] 를 만들 수 있습니다. 이 중 가장 큰 숫자는 94 입니다.

문자열 형식으로 숫자 number와 제거할 수의 개수 k가 solution 함수의 매개변수로 주어집니다. number에서 k 개의 수를 제거했을 때 만들 수 있는 수 중 가장 큰 숫자를 문자열 형태로 return 하도록 solution 함수를 완성하세요.

제한 조건
- number는 2자리 이상, 1,000,000자리 이하인 숫자입니다.
- k는 1 이상 number의 자릿수 미만인 자연수입니다.

입출력 예
number	k	return
"1924"	2	"94"
"1231234"	3	"3234"
"4177252841"	4	"775841"
```

---

### 풀이
```txt
가장 큰 수를 픽하는 기준이 있어야 한다.
맨 앞에 숫자가 가장 큰 숫자여야 한다.
그리고 자릿수가 내려갈 수록 순차적으로 내려가면 된다.

방법
stack으로 해결할 수 있다.
1. number를 순회한다.
2. 
  1) while 에서 k > 0, stack이 비어있지 않고, stack에 마지막에 추가된 숫자가 현재 숫자보다 작은 경우 -> stack.pop(), k -= 1
k -= 1은 문자 하나가 제거되었다는 뜻이다. k = 0으로 되는 순간 제거할 수 있는 숫자 개수를 모두 제거하였다는 뜻. 이렇게 해서 순서대로 제거하고 더 큰 숫자가 앞에 오게 할 수 있다.
  2) 숫자를 stack에 append
3. 내림차순 같은 경우 while 문에서 제거를 하지 못하므로 남은 k만큼 추가로 제거해야 하는데, 뒤쪽이 낮은 수가 몰려있어서 stack[:-k] 만큼으로 슬라이싱하면 된다.
```

````python
def solution(number, k):
    lsStack = []

    for i in number:
        while k > 0 and len(lsStack) > 0 and lsStack[-1] < i:
            lsStack.pop()
            k -= 1

        lsStack.append(i)

    if k > 0:
        lsStack = lsStack[:-k]

    return "".join(lsStack)


number = "1924"
k = 2
result = solution(number, k)
print(result)

number = "1231234"
k = 3
result = solution(number, k)
print(result)

number = "4177252841"
k = 4
result = solution(number, k)
print(result)
````

### 풀이
```txt
brute force 방법 -> 답은 맞는데 시간 초과
1. 먼저 숫자를 문자열로 각 한 자릿수로 나눈다. 
2. k만큼 숫자를 제거해야 하는데, 남는 개수는 len(number) - k이다. 그럼 combination으로 이만큼 돌리면 k개 만큼 제거된 숫자만 구성할 수 있다.
* combination은 오름차순으로 인덱스를 구성해준다.
* permutation은 순서를 보장하지 않는다. (0, 1), (1, 2)도 나오는데, (1, 0)도 나온다.
4. 정렬 후 마지막 숫자 return

```

````python
from itertools import combinations

def solution(number, k):
    answer = ''
    number = list(number)
    nLen = len(number)
    lsNumber_new = []
    for i in combinations(number, nLen - k) : 
        sNum = ""
        for j in i : 
            sNum += j
        lsNumber_new.append(int(sNum))
    lsNumber_new = sorted(lsNumber_new)
    #print(lsNumber_new)
    answer = str(lsNumber_new[-1])
    
    return answer


number = "1924"
k = 2
result = solution(number, k)
print(result)
````
