```txt
문제 설명
전화번호부에 적힌 전화번호 중, 한 번호가 다른 번호의 접두어인 경우가 있는지 확인하려 합니다.
전화번호가 다음과 같을 경우, 구조대 전화번호는 영석이의 전화번호의 접두사입니다.

- 구조대 : 119
- 박준영 : 97 674 223
- 지영석 : 11 9552 4421
전화번호부에 적힌 전화번호를 담은 배열 phone_book 이 solution 함수의 매개변수로 주어질 때, 어떤 번호가 다른 번호의 접두어인 경우가 있으면 false를 그렇지 않으면 true를 return 하도록 solution 함수를 작성해주세요.

제한 사항
- phone_book의 길이는 1 이상 1,000,000 이하입니다.
  - 각 전화번호의 길이는 1 이상 20 이하입니다.
  - 같은 전화번호가 중복해서 들어있지 않습니다.

입출력 예제
phone_book	return
["119", "97674223", "1195524421"]	false
["123","456","789"]	true
["12","123","1235","567","88"]	false

입출력 예 설명
입출력 예 #1
앞에서 설명한 예와 같습니다.

입출력 예 #2
한 번호가 다른 번호의 접두사인 경우가 없으므로, 답은 true입니다.

입출력 예 #3
첫 번째 전화번호, “12”가 두 번째 전화번호 “123”의 접두사입니다. 따라서 답은 false입니다.
```


````python
# 풀이 코드가 최대 O(n^2)이라서 별로 안 좋은 것 같아 다른 방법을 알아보았다.
# 문자열로 정렬을 하면 ["119", "1195524421", "97674223"]와 같이 접두어가 앞에 먼저 온다.
# 그래서 i, i + 1만 비교하면 된다.

# 시간복잡도 정렬 O(nlogn)

def solution(phone_book):
    phone_book.sort()

    for i in range(len(phone_book) - 1):
        sCur = phone_book[i]
        sNext = phone_book[i + 1]

        if sNext.startswith(sCur):
            return False

    return True
````

````python
# 주어진 번호 전체가 접두번호이다. 앞에서 일부의 번호가 접두 번호가 아니다.
# 전화 번호를 dict에 저장 -> 다른 전화 번호에서 접두가 없으면 넘어가고, 있으면 answer = False로 리턴

# 시간복잡도 : O(n * m)
# n : phone_book
# m : 번화번호 길이

def solution(phone_book):
    if len(phone_book) == 1 : 
        return True
    answer = True
    dic = {}
    # 접두번호 저장
    for i in phone_book : 
        dic[i] = 1

    for i in phone_book : 
        # 마지막 인덱스는 전체 번호이므로 생략
        for j in range(0, len(i) - 1) :  
            if i[:j + 1] in dic : 
                answer = False
                break
        if answer == False : 
            break

    return answer

phone_book = ["119", "97674223", "1195524421"]
result = solution(phone_book)
print(result)

phone_book = ["123","456","789"]
result = solution(phone_book)
print(result)

phone_book = ["12","123","1235","567","88"]
result = solution(phone_book)
print(result)

````
