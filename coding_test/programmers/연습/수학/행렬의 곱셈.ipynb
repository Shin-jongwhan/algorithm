```txt
문제 설명
2차원 행렬 arr1과 arr2를 입력받아, arr1에 arr2를 곱한 결과를 반환하는 함수, solution을 완성해주세요.

제한 조건
- 행렬 arr1, arr2의 행과 열의 길이는 2 이상 100 이하입니다.
- 행렬 arr1, arr2의 원소는 -10 이상 20 이하인 자연수입니다.
- 곱할 수 있는 배열만 주어집니다.

입출력 예
arr1	arr2	return
[[1, 4], [3, 2], [4, 1]]	[[3, 3], [3, 3]]	[[15, 15], [15, 15], [15, 15]]
[[2, 3, 2], [4, 2, 4], [3, 1, 4]]	[[5, 4, 3], [2, 4, 1], [3, 1, 1]]	[[22, 22, 11], [36, 28, 18], [29, 20, 14]]
```


````python
# 행렬의 곱셈은 곱하는 두 행렬의 형상이 특정 규칙으로 일치해야 한다.
# 행렬 A, B
# 각 형상은 A_h, A_w, B_h, B_w 라고 하자.
# 그러면 A_w = B_h이어야 한다.
# 리턴하는 형상은 A_h, B_w 이다.

# lsNew를 형상에 맞춰 0 행렬으로 초기화한다.
# A_h를 i로 접근, A_w를 j로 접근, B_w를 k로 접근하고 순서대로 for문을 만들면 된다.
# lsNew[i][k] += A[i][j] * B[j][k]
# 종종 for 문을 어떻게 중첩해야 할지 헷갈리는데, 형상을 기억하면 만들기 쉽다.

# 시간복잡도 : O(n*m*k)

def solution(arr1, arr2):
    answer = [ [0] * len(arr2[0]) for i in range(0, len(arr1)) ]
    for i in range(0, len(arr1)) :
        for j in range(0, len(arr1[0])) : 
            for k in range(0, len(arr2[0])) : 
                answer[i][k] += arr1[i][j] * arr2[j][k]
                
    return answer

arr1 = [[1, 4], [3, 2], [4, 1]]
arr2 = [[3, 3], [3, 3]]
result = solution(arr1, arr2)
print(result)

arr1 = [[2, 3, 2], [4, 2, 4], [3, 1, 4]]
arr2 = [[5, 4, 3], [2, 4, 1], [3, 1, 1]]
result = solution(arr1, arr2)
print(result)

````
