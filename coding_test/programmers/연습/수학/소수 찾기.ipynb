### 문제
#### https://school.programmers.co.kr/learn/courses/30/lessons/42839

```txt
문제 설명
한자리 숫자가 적힌 종이 조각이 흩어져있습니다. 흩어진 종이 조각을 붙여 소수를 몇 개 만들 수 있는지 알아내려 합니다.

각 종이 조각에 적힌 숫자가 적힌 문자열 numbers가 주어졌을 때, 종이 조각으로 만들 수 있는 소수가 몇 개인지 return 하도록 solution 함수를 완성해주세요.

제한사항
- numbers는 길이 1 이상 7 이하인 문자열입니다.
- numbers는 0~9까지 숫자만으로 이루어져 있습니다.
- "013"은 0, 1, 3 숫자가 적힌 종이 조각이 흩어져있다는 의미입니다.

입출력 예
numbers	return
"17"	3
"011"	2

입출력 예 설명
예제 #1
[1, 7]으로는 소수 [7, 17, 71]를 만들 수 있습니다.

예제 #2
[0, 1, 1]으로는 소수 [11, 101]를 만들 수 있습니다.
- 11과 011은 같은 숫자로 취급합니다.

```


---

### 풀이
```txt
1. 소수
어떤 수가 소수인지 다시 한 번 알아보자면 소수란 '1보다 큰 자연수 중에서, 약수가 1과 자기 자신뿐'인 수이다.
어떤 수 n = a * b로 된다고 했을 때, a, b는 루트(n) 보다는 클 수 없다. 그래서 2 이상, 루트(n) 이하인 수에서 다른 약수가 있는지 찾으면 된다.
여기서 또 효율을 높이자면, 짝수인 수는 항상 2가 다른 약수가 되기 때문에 n % 2 == 0인 수는 바로 False로 하면 되고, 홀수만 검사하면 된다.

2. numbers의 조합
  1) 마지막 수가 홀수인 것에서, 나머지 수의 조합으로 구하면 된다.
  2) 그리고 '선택하지 않음' 이라는 선택도 있다. 모두 다 선택하지 않음이라는 경우만 없애면 모든 경우의 수를 구할 수 있다.
숫자를 1개만 선택하는 경우, 2개, 3개, ... n개 선택하는 경우로 나눠야 한다.
이 경우 itertools에 permutations를 이용해서 구할 수 있다.

방법
1. numbers 안에 모든 수에 대해 홀수만 구성해서 1개만 뽑아서 일단 소수인지 검사하고, [ [[홀수 A를 제외한 전체 수], 홀수 A], [[홀수 B를 제외한 전체 수], 홀수 B], ... ]로 일단 저장한다.
* 1자리 소수는 2, 3, 5, 7만 있어서 이것만 검사하면 된다.
2. 1번에서 구한 홀수 A, B ...를 그걸 마지막 수로 하고, [홀수 A를 제외한 전체 수] ... 를 permutation을 돌려서 수를 구한다. 단, str 형태로 concat -> int로 바꿔서 해야 앞에 0을 제외할 수 있음.
3. 그다음 set에 저장.
4. set에 저장된 것을 for loop을 돌려서 is_prime() 호출 -> 소수면 answer + 1


```

````python
from itertools import permutations

def is_prime(nNum):
    if nNum < 2:
        return False
    if nNum == 2:
        return True
    if nNum % 2 == 0:
        return False

    nDiv = 3
    while nDiv * nDiv <= nNum:
        if nNum % nDiv == 0:
            return False
        nDiv += 2

    return True


def solution(numbers):
    answer = 0

    lsDigits = list(numbers)

    # 3) set에 저장 (한 자리 수도 같이 넣어서 중복 제거를 통일)
    stNumbers = set()

    # 1) 마지막 후보(홀수) 케이스 저장 + 1개만 픽한 수(홀수/2)는 stNumbers에 저장
    lsOdd_last_cases = []

    for nIdx in range(0, len(lsDigits)):
        sPick = lsDigits[nIdx]
        nPick = ord(sPick) - ord('0')

        if (nPick % 2 == 1) or (nPick == 2):
            stNumbers.add(nPick)

        if nPick % 2 == 1:
            lsRest = lsDigits[:nIdx] + lsDigits[nIdx + 1:]
            lsOdd_last_cases.append([lsRest, sPick])

    # 2) 마지막 홀수 고정 + 나머지 permutations로 수 생성 -> set에 저장
    for lsRest, sLast in lsOdd_last_cases:
        for nLen in range(0, len(lsRest) + 1):
            for tpPerm in permutations(lsRest, nLen):
                sFull = "".join(tpPerm) + sLast
                stNumbers.add(int(sFull))  # "011"과 "11"을 같은 수로 처리

    # 4) set에 저장된 것을 for loop으로 is_prime() 호출 -> 소수면 answer + 1
    for nNum in stNumbers:
        if is_prime(nNum):
            answer += 1

    return answer


numbers = "17"
result = solution(numbers)
print(result)

numbers = "011"
result = solution(numbers)
print(result)
````
