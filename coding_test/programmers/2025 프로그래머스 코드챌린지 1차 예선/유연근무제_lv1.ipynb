````python
def solution(schedules, timelogs, startday):
    answer = 0
    
    # 한 번도 늦지 않아야 보상
    # 토, 일 index 구분 필요
    # timelogs[i][j] - (schedules[i] + 10분) <= 0인 조건을 모두 만족해야 보상
    # schedules, timelogs의 길이는 같고, 첫 번째 차원 idx는 똑같은 직원의 정보임.
    
    # 시, 분을 구분해야 함.
    ## +10분을 할 때 50 이상이면 시간 + 1, 분은 맨 앞자리 수로 함. 50 미만이면 그대로 +10분.
    ## 700 ≤ schedules[i] ≤ 1100 이므로 23시가 넘을 가정은 안 해도 됨. 만약 넘는 경우라면 % 24로 연산하면 됨.
    if startday == 7 : 
        nSat_idx = 6
    else : 
        nSat_idx = 6 - startday
    nSun_idx = 7 - startday
        
    for i in range(0, len(schedules)) : 
        nAdjust_time = 0
        nHour_adjust = schedules[i] // 100
        nMinute_adjust = (schedules[i] % 100) + 10
        if nMinute_adjust >= 60 : 
            nAdjust_time = int("{0}{1:02d}".format(nHour_adjust + 1, nMinute_adjust % 60))
        else : 
            nAdjust_time = schedules[i] + 10

        # if 문을 이렇게 해도 됨. 그런데 이렇게하면 str -> int 연산을 매번 하므로 최적화 손해.
        # 다른 방법 : 시, 분을 나눈 변수끼리 그대로 비교해도 됨
        # nHour_adjust += nMinute_adjust // 60
        # nMinute_adjust = nMinute_adjust % 60
        # nAdjust_time = int("{0}{1:02d}".format(nHour_adjust, nMinute_adjust))
            
        blReward = True
        for j in range(0, len(timelogs[i])) : 
            # 토, 일 스킵
            if nSat_idx == j or nSun_idx == j :
                continue
            
            if timelogs[i][j] - nAdjust_time > 0 :
                blReward = False
                break
        if blReward == True : 
            answer += 1
    
    return answer
````
