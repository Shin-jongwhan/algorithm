```txt
문제 설명
JadenCase란 모든 단어의 첫 문자가 대문자이고, 그 외의 알파벳은 소문자인 문자열입니다. 단, 첫 문자가 알파벳이 아닐 때에는 이어지는 알파벳은 소문자로 쓰면 됩니다. (첫 번째 입출력 예 참고)
문자열 s가 주어졌을 때, s를 JadenCase로 바꾼 문자열을 리턴하는 함수, solution을 완성해주세요.

제한 조건
s는 길이 1 이상 200 이하인 문자열입니다.
s는 알파벳과 숫자, 공백문자(" ")로 이루어져 있습니다.
숫자는 단어의 첫 문자로만 나옵니다.
숫자로만 이루어진 단어는 없습니다.
공백문자가 연속해서 나올 수 있습니다.
입출력 예
s	return
"3people unFollowed me"	"3people Unfollowed Me"
"for the last week"	"For The Last Week"

```


````python
# 1. 모두 소문자로 만들고 문자열을 ' '로 자르기. 
# 2. 아스키코드로 a~z 범위 내에 있으면 첫 문자를 .upper()
# 3. 합치기

# 시간복잡도 : O(n)
# n = 띄어쓰기로 나눈 s 길이

def solution(s):
    if s == "" : 
        return s
    answer = ''
    nAsc_min_lower = ord('a')     # 97
    nAsc_max_lower = ord('z')     # 122
    
    lsSplit = s.lower().split(" ")
    for i in range(0, len(lsSplit)) : 
        if len(lsSplit[i]) == 0 : 
            continue
        if nAsc_min_lower <= ord(lsSplit[i][0]) and ord(lsSplit[i][0]) <= nAsc_max_lower : 
            lsSplit[i] = lsSplit[i][0].upper() + lsSplit[i][1:]
    
    answer = " ".join(lsSplit)
    
    return answer


# 답 "3people Unfollowed Me"
s = "3people unFollowed me"
result = solution(s)
print(result)

# 답 "For The Last Week"
s = "for the last week"
result = solution(s)
print(result)
````
