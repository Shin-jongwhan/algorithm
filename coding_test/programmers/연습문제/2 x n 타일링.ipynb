# DP(dynamic programming) 방법


```txt
세로 배치와 가로 배치는 서로 겹칠 수 없다는 것을 알고 있으면 좋은 힌트가 된다.
먼저 n = 1일 때는 세로 배치 1개만 들어갈 수 있다. -> 경우의 수 1개
n = 2일 때는 세로 배치 2개 / 가로 배치 2개만 들어갈 수 있다. -> 경우의 수 2개

n = 3 일 때를 생각해보자.
먼저 맨 왼쪽에 세로 배치 1개를 배치해보고 나머지 경우의 수를 생각해보자. -> 그러면 n = 2일 때의 경우의 수와 같다는 것을 알 수 있다.
맨 왼쪽에 가로 배치 2개를 배치해보고 나머지 경우의 수를 생각해보자. -> 그러면 n = 1일 때의 경우의 수와 같다는 것을 알 수 있다.

- 1번째 세로칸까지 칠해진 경우 -> n = 2일 때의 경우의 수를 고려
┌──────┬──────┬──────┐
│  █   │      │      │
├──────┼──────┼──────┤
│  █   │      │      │
└──────┴──────┴──────┘

- 2번째 세로칸까지 칠해진 경우 -> n = 1일 때의 경우의 수를 고려
┌──────┬──────┬──────┐
│  █   │  █   │      │
├──────┼──────┼──────┤
│  █   │  █   │      │
└──────┴──────┴──────┘

따라서 n = 3일 때의 경우의 수는 (n = 2일 때 경우의 수) + (n = 1일때 경우의 수)


n = 4일 때를 생각해보자.
- 1번째 세로칸까지 칠해진 경우 -> n = 3일 때의 경우의 수를 고려
┌──────┬──────┬──────┬──────┐
│  █   │      │      │      │
├──────┼──────┼──────┼──────┤
│  █   │      │      │      │
└──────┴──────┴──────┴──────┘

- 2번째 세로칸까지 칠해진 경우 -> n = 2일 때의 경우의 수를 고려
┌──────┬──────┬──────┬──────┐
│  █   │  █   │      │      │
├──────┼──────┼──────┼──────┤
│  █   │  █   │      │      │
└──────┴──────┴──────┴──────┘

따라서 n = 4일 때의 경우의 수는 (n = 3일 때 경우의 수) + (n = 2일때 경우의 수)

계속 가서 n = n일 때의 경우의 수 = (n - 2일 때 경우의 수) + (n - 1일때 경우의 수)

```


````python
# DP(dynamic programming) 방법

# 시간복잡도 : O(n)

def solution(n):
    if n <= 2:
        return n
    
    nMod = 1000000007       # 제한 조건. 이걸로 나눈 나머지를 리턴

    nAnswer = 0
    nPrev_1 = 1     # n = 1일 때 경우의 수로 초기화
    nPrev_2 = 2     # n = 2일 때 경우의 수로 초기화
    for i in range(3, n + 1) : 
        nTmp = nPrev_2
        nPrev_2 = nPrev_1 + nPrev_2
        nPrev_1 = nTmp

    nAnswer = nPrev_2 % nMod

    return nAnswer


# 답 5
n = 4
result = solution(n)
print(result)

# 답 3
n = 3
result = solution(n)
print(result)

# 답 8
n = 5
result = solution(n)
print(result)

# 답 13
n = 6
result = solution(n)
print(result)
````

````python
# 가로 배치를 고정하고, 나머지가 세로 배치를 하면 된다.

# 가로 길이는 2 이다. 가로 배치를 H라고 하자. 
# 세로 배치는 V라고 하자.
# H가 k개 있으면, 길이는 2k
# 남은 길이는 n - 2k이고, 이 남은 자리에 세로 배치를 끼워넣으면 된다.
# 이에 대한 조합식을 모두 더하면 답이 된다.
# H 개수 : k개
# V 개수 : n - 2k개
# 전체 블록 개수 : H + V = n - k개
# 이게 블록이 들어갈 수 있는 자리 수가 된다.
# 그러면 k개의 가로 배치 H가 n - k 자리 수에서 나올 수 있는 조합은 math.comb(n - k, k) 이다.
# 이걸 H가 없는 0개부터 k 최대 개수인 n // 2 까지 모두 더하면 된다.


# 검증 통과, 효율성 테스트 실패.
# 조합식으로 풀면 시간복잡도가 너무 커서 안 된다. 다른 방법으로 해야 함.

import math

def solution(n):
    if n <= 2:
        return n

    nAnswer = 0
    nMax_h = n // 2  # H(가로 블록) 최대 개수

    for nH in range(0, nMax_h + 1):
        nAnswer += math.comb(n - nH, nH)

    return nAnswer

# 답 5
n = 4
result = solution(n)
print(result)

# 답 3
n = 3
result = solution(n)
print(result)

# 답 8
n = 5
result = solution(n)
print(result)

# 답 13
n = 6
result = solution(n)
print(result)
````
