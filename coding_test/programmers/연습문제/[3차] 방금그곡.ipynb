### 문제
#### https://school.programmers.co.kr/learn/courses/30/lessons/17683

```txt
문제 설명
방금그곡
라디오를 자주 듣는 네오는 라디오에서 방금 나왔던 음악이 무슨 음악인지 궁금해질 때가 많다. 그럴 때 네오는 다음 포털의 '방금그곡' 서비스를 이용하곤 한다. 방금그곡에서는 TV, 라디오 등에서 나온 음악에 관해 제목 등의 정보를 제공하는 서비스이다.

네오는 자신이 기억한 멜로디를 가지고 방금그곡을 이용해 음악을 찾는다. 그런데 라디오 방송에서는 한 음악을 반복해서 재생할 때도 있어서 네오가 기억하고 있는 멜로디는 음악 끝부분과 처음 부분이 이어서 재생된 멜로디일 수도 있다. 반대로, 한 음악을 중간에 끊을 경우 원본 음악에는 네오가 기억한 멜로디가 들어있다 해도 그 곡이 네오가 들은 곡이 아닐 수도 있다. 그렇기 때문에 네오는 기억한 멜로디를 재생 시간과 제공된 악보를 직접 보면서 비교하려고 한다. 다음과 같은 가정을 할 때 네오가 찾으려는 음악의 제목을 구하여라.

- 방금그곡 서비스에서는 음악 제목, 재생이 시작되고 끝난 시각, 악보를 제공한다.
- 네오가 기억한 멜로디와 악보에 사용되는 음은 C, C#, D, D#, E, F, F#, G, G#, A, A#, B 12개이다.
- 각 음은 1분에 1개씩 재생된다. 음악은 반드시 처음부터 재생되며 음악 길이보다 재생된 시간이 길 때는 음악이 끊김 없이 처음부터 반복해서 재생된다. 음악 길이보다 재생된 시간이 짧을 때는 처음부터 재생 시간만큼만 재생된다.
- 음악이 00:00를 넘겨서까지 재생되는 일은 없다.
- 조건이 일치하는 음악이 여러 개일 때에는 라디오에서 재생된 시간이 제일 긴 음악 제목을 반환한다. 재생된 시간도 같을 경우 먼저 입력된 음악 제목을 반환한다.
- 조건이 일치하는 음악이 없을 때에는 “(None)”을 반환한다.

입력 형식
입력으로 네오가 기억한 멜로디를 담은 문자열 m과 방송된 곡의 정보를 담고 있는 배열 musicinfos가 주어진다.

- m은 음 1개 이상 1439개 이하로 구성되어 있다.
- musicinfos는 100개 이하의 곡 정보를 담고 있는 배열로, 각각의 곡 정보는 음악이 시작한 시각, 끝난 시각, 음악 제목, 악보 정보가 ','로 구분된 문자열이다.
- 음악의 시작 시각과 끝난 시각은 24시간 HH:MM 형식이다.
- 음악 제목은 ',' 이외의 출력 가능한 문자로 표현된 길이 1 이상 64 이하의 문자열이다.
- 악보 정보는 음 1개 이상 1439개 이하로 구성되어 있다.

출력 형식
- 조건과 일치하는 음악 제목을 출력한다.

입출력 예시
m	musicinfos	answer
"ABCDEFG"	["12:00,12:14,HELLO,CDEFGAB", "13:00,13:05,WORLD,ABCDEF"]	"HELLO"
"CC#BCC#BCC#BCC#B"	["03:00,03:30,FOO,CC#B", "04:00,04:08,BAR,CC#BCC#BCC#B"]	"FOO"
"ABC"	["12:00,12:14,HELLO,C#DEFGAB", "13:00,13:05,WORLD,ABCDEF"]	"WORLD"

설명
첫 번째 예시에서 HELLO는 길이가 7분이지만 12:00부터 12:14까지 재생되었으므로 실제로 CDEFGABCDEFGAB로 재생되었고, 이 중에 기억한 멜로디인 ABCDEFG가 들어있다.
세 번째 예시에서 HELLO는 C#DEFGABC#DEFGAB로, WORLD는 ABCDE로 재생되었다. HELLO 안에 있는 ABC#은 기억한 멜로디인 ABC와 일치하지 않고, WORLD 안에 있는 ABC가 기억한 멜로디와 일치한다.
```

### 풀이
```txt
문제가 좀 난해했다. 
문제를 이해하는 것 자체도 어려웠고, 부분 문자열 매칭을 해야 하나 명확하지 않았고, 조건이 많아서 혼동이 많이 되었다. 

음의 기억이 음악의 끝과 처음이 맞닿아서 기억할 수도 있다고 했으므로 이에 대한 해결책으로
- 음악 음 전체를 2개 이어붙이고 그 안에서 조회가 되는지 검색하면 되지 않을까 생각했다. -> 괜찮지만 시간복잡도가 문제.
- 또는 각 음의 순서를 dict 트리로 조회하는 구조를 만들면 되겠다 생각했다. -> 구현을 어떻게?
그런데 중간에 있는 음이 있는지 조회하려면 이건 prefix가 아니라 suffix tree로 만들어야 하기 때문에 구현이 복잡할 것 같았다. 
그런데 다시 읽어보니 조건이 부분문자열이 매칭이 되는지 찾는 건 아니었다. -> 그냥 주어진 멜로디 문자열 m 길이 만큼 일치하는지만 판단하면 되는 문제.

C# 같은 건 한 문자로 치환해서 찾는 게 좋다. 안 그러면 에러케이스가 발생한다.
예를 들어 ABC인데 음 S가 ABC# 이라면 ABC가 매칭이 된다.
이것때문에 오해할 수 있는데, C#DEFGABC#DEFGAB은 16개 길이라고 16분이 재생된 거 아닌가 하는 착각이 드는데, C#이 2개 들어가 있어서 실제로는 14분 재생이 맞다.

방법 
1. # 붙어있는 음 소문자로 변환
2. 각 음악 재생 시간을 고려하여 음을 이어붙이기. 시간 -> 분 변환 함수를 만들고 분 길이로 이어붙이기.
3. 기억 음 m 매칭이 있는지 확인. 전체가 들어있는지 확인하는 것 뿐이므로 그냥 in 비교로 하면 됨.
4. 있으면 후보로 음악과 가장 긴 재생 시간을 변수로 저장한다. 변수는 각각 빈 문자열 "(None)", 0으로 저장
5. 해당 제목 리턴

시간복잡도 : 최악 O(n * m)
n : 재생 시간에 대한 음악 음 전체 길이
m : 음악 개수
```

````python
def convert_sharp(sMelody):
    # 음 단위로 스캔하면서, '#"가 붙은 음은 소문자 1글자로 바꾼다.
    # 예) "C#DEFGAB" -> "cDEFGAB"
    dicSharp = {
        "C": "c",
        "D": "d",
        "E": "e",
        "F": "f",
        "G": "g",
        "A": "a",
        "B": "b",
    }

    lsOut = []
    i = 0
    while i < len(sMelody):
        if i + 1 < len(sMelody) and sMelody[i + 1] == "#":
            lsOut.append(dicSharp.get(sMelody[i], sMelody[i].lower()))
            i += 2
        else:
            lsOut.append(sMelody[i])
            i += 1

    return "".join(lsOut)


def get_play_time(sStart, sEnd):
    # "HH:MM" -> 분 단위로 변환 후 차이 계산
    nStart_h = int(sStart[0:2])
    nStart_m = int(sStart[3:5])
    nEnd_h = int(sEnd[0:2])
    nEnd_m = int(sEnd[3:5])

    nStart_total = nStart_h * 60 + nStart_m
    nEnd_total = nEnd_h * 60 + nEnd_m
    return nEnd_total - nStart_total


def solution(m, musicinfos):
    # 기억 멜로디도 샾 변환
    sM = convert_sharp(m)

    # 정답 후보(제목/재생시간) 초기화
    sBest_title = "(None)"
    nBest_time = 0

    for i in range(0, len(musicinfos)):
        sInfo = musicinfos[i]
        lsParts = sInfo.split(",")

        sStart = lsParts[0]
        sEnd = lsParts[1]
        sTitle = lsParts[2]
        sScore_raw = lsParts[3]

        # 재생 시간 계산 + 악보 변환
        nPlay_time = get_play_time(sStart, sEnd)
        sScore = convert_sharp(sScore_raw)

        # 재생 시간이 0이면 스킵(문제 조건상 거의 없지만 안전 처리)
        if nPlay_time <= 0:
            continue
        
        # 기억 멜로디가 재생시간보다 길면 포함될 수 없음
        if len(sM) > nPlay_time:
            continue

        # 재생 시간만큼 실제 재생 멜로디 만들기(반복 + 절단)
        nScore_len = len(sScore)
        nRepeat = nPlay_time // nScore_len + 1
        sPlayed = (sScore * nRepeat)[:nPlay_time]

        # 포함 여부 확인 (전체 문자열 매칭만 필요)
        if sM in sPlayed:
            # 더 긴 재생시간이면 갱신
            # 같으면 먼저 입력된 곡 유지해야 하므로 갱신하지 않음
            if nPlay_time > nBest_time:
                nBest_time = nPlay_time
                sBest_title = sTitle

    # 제목 리턴
    return sBest_title


m = "ABCDEFG"
musicinfos = ["12:00,12:14,HELLO,CDEFGAB", "13:00,13:05,WORLD,ABCDEF"]
result = solution(m, musicinfos)
print(result)

m = "CC#BCC#BCC#BCC#B"
musicinfos = ["03:00,03:30,FOO,CC#B", "04:00,04:08,BAR,CC#BCC#BCC#B"]
result = solution(m, musicinfos)
print(result)

m = "ABC"
musicinfos = ["12:00,12:14,HELLO,C#DEFGAB", "13:00,13:05,WORLD,ABCDEF"]
result = solution(m, musicinfos)
print(result)
````
