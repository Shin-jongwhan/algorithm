```txt
문제 설명
N개의 마을로 이루어진 나라가 있습니다. 이 나라의 각 마을에는 1부터 N까지의 번호가 각각 하나씩 부여되어 있습니다. 각 마을은 양방향으로 통행할 수 있는 도로로 연결되어 있는데, 서로 다른 마을 간에 이동할 때는 이 도로를 지나야 합니다. 도로를 지날 때 걸리는 시간은 도로별로 다릅니다. 현재 1번 마을에 있는 음식점에서 각 마을로 음식 배달을 하려고 합니다. 각 마을로부터 음식 주문을 받으려고 하는데, N개의 마을 중에서 K 시간 이하로 배달이 가능한 마을에서만 주문을 받으려고 합니다. 다음은 N = 5, K = 3인 경우의 예시입니다.
```

```txt
위 그림에서 1번 마을에 있는 음식점은 [1, 2, 4, 5] 번 마을까지는 3 이하의 시간에 배달할 수 있습니다. 그러나 3번 마을까지는 3시간 이내로 배달할 수 있는 경로가 없으므로 3번 마을에서는 주문을 받지 않습니다. 따라서 1번 마을에 있는 음식점이 배달 주문을 받을 수 있는 마을은 4개가 됩니다.
마을의 개수 N, 각 마을을 연결하는 도로의 정보 road, 음식 배달이 가능한 시간 K가 매개변수로 주어질 때, 음식 주문을 받을 수 있는 마을의 개수를 return 하도록 solution 함수를 완성해주세요.

제한사항
- 마을의 개수 N은 1 이상 50 이하의 자연수입니다.
- road의 길이(도로 정보의 개수)는 1 이상 2,000 이하입니다.
- road의 각 원소는 마을을 연결하고 있는 각 도로의 정보를 나타냅니다.
- road는 길이가 3인 배열이며, 순서대로 (a, b, c)를 나타냅니다.
  - a, b(1 ≤ a, b ≤ N, a != b)는 도로가 연결하는 두 마을의 번호이며, c(1 ≤ c ≤ 10,000, c는 자연수)는 도로를 지나는데 걸리는 시간입니다.
  - 두 마을 a, b를 연결하는 도로는 여러 개가 있을 수 있습니다.
  - 한 도로의 정보가 여러 번 중복해서 주어지지 않습니다.
- K는 음식 배달이 가능한 시간을 나타내며, 1 이상 500,000 이하입니다.
- 임의의 두 마을간에 항상 이동 가능한 경로가 존재합니다.
- 1번 마을에 있는 음식점이 K 이하의 시간에 배달이 가능한 마을의 개수를 return 하면 됩니다.

입출력 예
```

```txt
1번 마을에서 배달에 4시간 이하가 걸리는 마을은 [1, 2, 3, 5] 4개이므로 4를 return 합니다.
```




````python
import heapq

# 다익스트라 방법
# (최소 누적 거리, 노드)로 heap으로 정렬 -> pop 하고 해당 누적 거리가 기록된 거리보다 작다면 노드 방문 -> 반복 ... 을 하는 것이다.

# 1. dist[1]=0, 나머지 INF
# 2. 힙에 (0,1) push
# 3. 힙에서 (거리, 노드) pop
# 4. pop한 거리값이 dist[노드]랑 다르면(낡은 후보면) skip
# 5. 그 노드의 모든 이웃에 대해
# 6. 새거리 = pop거리 + 간선비용
# 7. 새거리 < dist[이웃]이면 dist[이웃]=새거리로 갱신하고 힙에 push
# 8. 힙이 비면 종료
# 9. dist[i] <= K 개수 세서 return

def solution(N, road, K):
    answer = 0 
    
    dicRoad = {}

    # 그래프 구성: dicRoad[a][b] = 최소 거리
    for i in road:
        nA = i[0]
        nB = i[1]
        nC = i[2]

        if nA not in dicRoad:
            dicRoad[nA] = {}
        if nB not in dicRoad:
            dicRoad[nB] = {}

        if nB not in dicRoad[nA]:
            dicRoad[nA][nB] = nC
        else:
            dicRoad[nA][nB] = min(dicRoad[nA][nB], nC)

        if nA not in dicRoad[nB]:
            dicRoad[nB][nA] = nC
        else:
            dicRoad[nB][nA] = min(dicRoad[nB][nA], nC)
            
    nInf = 10**10
    lsDist = [nInf] * (N + 1)       # 노드가 1부터 시작하므로 인덱스 맞추기 위함
    lsDist[1] = 0       # 시작 위치는 거리가 0임
    
    lsHeap = []
    heapq.heappush(lsHeap, (0, 1))
    
    while lsHeap : 
        nCur_dist, nNode = heapq.heappop(lsHeap)
        
        if nCur_dist > lsDist[nNode] : 
            continue
        
        for i in dicRoad[nNode] : 
            nDist_new = nCur_dist + dicRoad[nNode][i]
            if nDist_new < lsDist[i] : 
                lsDist[i] = nDist_new
                heapq.heappush(lsHeap, (nDist_new, i))
                
    print(lsDist)
    for i in range(1, len(lsDist)) : 
        if lsDist[i] <= K : 
            answer += 1
    
    return answer


N = 5
road = [[1,2,1],[2,3,3],[5,2,2],[1,4,2],[5,3,1],[5,4,2]]
K = 3
result = solution(N, road, K)
print(result)

N = 6
road = [[1,2,1],[1,3,2],[2,3,2],[3,4,3],[3,5,2],[3,5,3],[5,6,1]]
K = 4
result = solution(N, road, K)
print(result)
````

````python
# DFS로 백트랙킹 + best 거리를 저장하여 푸는 방법
# 알아보니 이렇게는 잘 안 푼다고 한다. n 수가 작아서 그렇지, 커지면 최적화에서 문제가 있을 수 있다.
# 이 문제는 다익스트라가 정석이라고 한다. 다익스트라 알고리즘이 가중치가 있는 간선에서 최단 거리를 찾는 알고리즘이어서 정확히 이 문제와 같다.

# dicRoad = {출발지 : {도착지 : 최소 이동거리}} 형태로 저장. 예시 그림 2번과 같이 같은 마을에 가는 길이 2개인데 걸리는 시간이 다르다면 최솟값으로 저장해야 함.
# 그러면 dicRoad[출발지].keys()가 연결된 마을임. 
# dicVisited_all = {}. 방문한 마을 전체. answer = len(dicVisited_all) 가 될 것임.

# 다음 마을 방문 함수 move(출발지, 이동거리)
# dicBest_dist[도착지]에 총 이동거리를 저장하여 가장 짧은 이동거리만 저장. 이 이동거리 이상인 루트로 가면 더 이상 방문 안 해도 됨
# dicRoad[출발지].keys()를 순회하며 방문
# 만약 전체 이동거리 + dicRoad[출발지][도착지] <= k이고, dicBest_dist에 키가 없거나 dicBest_dist[도착지] 보다 이동거리 작은 경우라면
#    visited_all[도착지] = True. 중복으로 적용해도 상관 없음.
#    new 이동거리 = 이동 거리 + dicRoad[출발지][도착지]
#    move(도착지, new 이동거리). 다음 마을로 이동
# else : return

def solution(N, road, K):
    dicRoad = {}

    # dicRoad 구성: dicRoad[a][b] = 최소 거리
    for lsEdge in road:
        nA = lsEdge[0]
        nB = lsEdge[1]
        nC = lsEdge[2]

        if nA not in dicRoad:
            dicRoad[nA] = {}
        if nB not in dicRoad:
            dicRoad[nB] = {}

        if nB not in dicRoad[nA]:
            dicRoad[nA][nB] = nC
        else:
            dicRoad[nA][nB] = min(dicRoad[nA][nB], nC)

        if nA not in dicRoad[nB]:
            dicRoad[nB][nA] = nC
        else:
            dicRoad[nB][nA] = min(dicRoad[nB][nA], nC)

    # 각 노드에 도달한 '현재까지 발견된 최단 거리'
    dicBest_dist = {1: 0}

    def move(nCur, nCur_dist):
        # 연결 간선이 없을 수도 있는 방어
        if nCur not in dicRoad:
            return

        for nNext in dicRoad[nCur]:
            nNext_dist = nCur_dist + dicRoad[nCur][nNext]
            if nNext_dist > K:
                continue

            # 더 짧게 도달한 경우에만 갱신/확장
            if nNext not in dicBest_dist or nNext_dist < dicBest_dist[nNext]:
                dicBest_dist[nNext] = nNext_dist
                move(nNext, nNext_dist)

    move(1, 0)

    # K 이하로 도달 가능한 노드 수
    nAnswer = len(dicBest_dist)

    return nAnswer

N = 5
road = [[1,2,1],[2,3,3],[5,2,2],[1,4,2],[5,3,1],[5,4,2]]
K = 3
#result = solution(N, road, K)
#print(result)

N = 6
road = [[1,2,1],[1,3,2],[2,3,2],[3,4,3],[3,5,2],[3,5,3],[5,6,1]]
K = 4
result = solution(N, road, K)
print(result)
````
