### 문제
#### https://school.programmers.co.kr/learn/courses/30/lessons/12946

```txt
문제 설명
하노이 탑(Tower of Hanoi)은 퍼즐의 일종입니다. 세 개의 기둥과 이 기동에 꽂을 수 있는 크기가 다양한 원판들이 있고, 퍼즐을 시작하기 전에는 한 기둥에 원판들이 작은 것이 위에 있도록 순서대로 쌓여 있습니다. 게임의 목적은 다음 두 가지 조건을 만족시키면서, 한 기둥에 꽂힌 원판들을 그 순서 그대로 다른 기둥으로 옮겨서 다시 쌓는 것입니다.

- 한 번에 하나의 원판만 옮길 수 있습니다.
- 큰 원판이 작은 원판 위에 있어서는 안됩니다.

하노이 탑의 세 개의 기둥을 왼쪽 부터 1번, 2번, 3번이라고 하겠습니다. 1번에는 n개의 원판이 있고 이 n개의 원판을 3번 원판으로 최소 횟수로 옮기려고 합니다.

1번 기둥에 있는 원판의 개수 n이 매개변수로 주어질 때, n개의 원판을 3번 원판으로 최소로 옮기는 방법을 return하는 solution를 완성해주세요.

제한사항
- n은 15이하의 자연수 입니다.

입출력 예
n	result
2	[ [1,2], [1,3], [2,3] ]
입출력 예 설명
입출력 예 #1
다음과 같이 옮길 수 있습니다.

```

### 방법에 대한 생각
```txt
이 문제는 푸는 방법을 몰라서 하루 종일 고민했다.

n = 1일때 옮겨보고, n = 2일때 옮겨보고, n = 3일 때 옮겨보고 ... 를 반복하다보면서 일정한 규칙을 확인하였다.
출발지 기둥을 A, 출발지 / 목적지 기둥아 아닌 기둥을 B, 도착해야 하는 목적지 기둥을 C 라고 하자.
n = 1일 때는 그냥 A -> C로 옮기면 됐다.
[[1,3]]
n = 2일 때는 2를 B로 옮기고 1을 C로, 그 다음 B에 있는 2를 C로 옮기면 됐다.
[[1,2], [1,3], [2, 3]]
n = 3일 때부터 생각을 해봐야 하는데, 일단 그려서 확인해보면 이렇게 된다.
[[1,3], [1,2], [3,2], [1,3], [2,1], [2,3], [1,3]]

이렇게 확인해보면 뭔가 규칙성이 없어보인다. 그런데 3개 기둥을 그려서 확인해보면 1~2개를 보조 기둥인 B로 옮기고, 3을 C로 옮기고, 그리고 보조 기둥 B에 있는 1~2를 C로 옮기는 과정에 있었다.

n = 4일 때도 마찬가지다. 1~3을 보조기둥인 B로, 4를 C로, B에 있는 1~3을 C로 옮긴다.
그러면 n = n일 때의 문제는 이러하다.
1. n - 1까지 전체를 보조 기둥 B로 옮긴다.
2. n을 목적지 기둥 C로 옮긴다.
3. 보조 기둥에 있는 n - 1 전체를 목적지 기둥 C로 옮긴다.

1, 3번을 잘 봐보자. 이것 자체도 하노이 탑을 옮기는 것과 같다.
1번에서는 보조 기둥 B가 목적지 기둥이 되며, 보조 기둥이 C가 된 것과 같다. 
3번에서는 보조 기둥 B가 출발지 기둥이 되며, A가 보조 기둥, C가 목적지 기둥이 된 것과 같다.

그럼 1, 3번은 결국 하노이 탑을 옮기는 행위는 같다는 걸 알았으므로 1번을 기준으로 좀 더 봐보자.
하노이 탑 n-1을 목적지 기둥으로 또 옮기려면 보조 기둥으로 n-2를 옮겨야 한다.
또 다시 n-2를 탑으로 쌓으려면 보조 기둥으로 n-3을 옮겨야 한다.
...
1이 될 때까지 반복한다. 1은 그냥 출발지 기둥에서 목적지 기둥으로 바로 옮길 수 있다.

그러면 파라미터는 이렇게 설계를 할 수 있다. n, 출발지 기둥 A, 보조 기둥 B, 목적지 기둥 C
그리고 함수를 hanoi()라고 하자. 그럼 hanoi(n, A, B, C) 로 표현할 수 있다.
그리고 hanoi() 함수는 이렇게 작동한다.
1. 1이면 출발지로 바로 감
2. n-1을 출발지 기둥을 A, 보조 기둥을 C, 목적지 기둥을 B로 한 hanoi(n-1, A, C, B)
3. n을 출발지 기둥에서 목적지 기둥으로 옮기는 경우의 수 (A, C)
4. n-1을 출발지 기둥을 B, 보조 기둥을 A, 목적지 기둥을 C로 한 hanoi(n-1, B, A, C)

```

````python
def hanoi(n, A, B, C, answer) : 
    # A 출발지 기둥
    # B 보조 기둥
    # C 목적지 기둥
    if n == 1 : 
        answer.append([A, C])
        return
    hanoi(n - 1, A, C, B, answer)
    answer.append([A, C])
    hanoi(n - 1, B, A, C, answer)
    

def solution(n):
    answer = []
    hanoi(n, 1, 2, 3, answer)
    
    return answer


n = 3
result = solution(n)
print(result)
````
