문제 설명
Finn은 요즘 수학공부에 빠져 있습니다. 수학 공부를 하던 Finn은 자연수 n을 연속한 자연수들로 표현 하는 방법이 여러개라는 사실을 알게 되었습니다. 예를들어 15는 다음과 같이 4가지로 표현 할 수 있습니다.

1 + 2 + 3 + 4 + 5 = 15
4 + 5 + 6 = 15
7 + 8 = 15
15 = 15
자연수 n이 매개변수로 주어질 때, 연속된 자연수들로 n을 표현하는 방법의 수를 return하는 solution를 완성해주세요.

제한사항
n은 10,000 이하의 자연수 입니다.

입출력 예
n	result
15	4

입출력 예 설명
입출력 예#1
문제의 예시와 같습니다.



````python
from collections import deque

# 연속된 자연수의 합의 최대값은 n/2 + 1보다 작다.
# 연속된 자연수의 합은 (시작 수 + 끝 수) * 길이 / 2로 구할 수 있다.
# 예를 들어 [1,2,3,4,5] 이면 (1 + 5) * 5 / 2 = 15이다.
# deque를 활용한다. dq라 하겠다. 그런데 딱히 dequeue를 안 쓰고, start, end 로 저장하고 길이를 end - start + 1로 구해도 상관은 없음(그게 더 좋음).
# if 추가하는 수 > 2/n + 1이라면 -> break
# if 자연수 합 == v 라면 -> answer += 1
# elif 자연수 합 < v 라면 -> 다음 수 추가
# elif 자연수 합 > v 라면 -> popleft()

# 시간복잡도 : O(m)
# m < (n / 2) + 1 이하 자연수

def solution(n):
    if n <= 1 : 
        return 1
    
    answer = 0
    dq = deque([1])
    val = 1
    while val < (n / 2) + 1 : 
        sum = (dq[0] + dq[-1]) * len(dq) // 2
        if sum < n : 
            val += 1
            dq.append(val)
        elif sum > n : 
            dq.popleft()
        elif sum == n : 
            answer += 1
            dq.popleft()
            val += 1
            dq.append(val)

    # 자기 자신
    answer += 1

    return answer


n = 15
result = solution(n)
print(result)

n = 2
result = solution(n)
print(result)

n = 3
result = solution(n)
print(result)

n = 4
result = solution(n)
print(result)

n = 10000
result = solution(n)
print(result)
````
