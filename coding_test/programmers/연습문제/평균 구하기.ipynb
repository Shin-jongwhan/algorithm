```txt
문제 설명
정수를 담고 있는 배열 arr의 평균값을 return하는 함수, solution을 완성해보세요.

제한사항
arr은 길이 1 이상, 100 이하인 배열입니다.
arr의 원소는 -10,000 이상 10,000 이하인 정수입니다.
입출력 예
arr	return
[1,2,3,4]	2.5
[5,5]	5
```


````python
# 배열 arr에서 평균을 리턴
# 다만, 예시에서 보면 정수는 정수로 리턴해야 한다.
# 3.0000000000000004와 같은 값을 잡기 위해 나눈 다음 a.is_integer() 로 판별하지는 말기
# 대신에 나머지를 nSum % nLen 으로 구해서 있으면 소수(float), 없으면 정수(integer)
## 나머지가 없어서 정수 리턴해야 할 때 : 정수 나눗셈 nSum // nLen 으로 리턴하면 정수를 리턴할 수 있다. 
## 나머지가 있다면 : nSum / nLen으로 그대로 리턴

def solution(arr):
    answer = 0
    nSum = sum(arr)
    nLen = len(arr)

    if nSum % nLen == 0:
        answer = nSum // nLen
    else : 
        answer = nSum / nLen
        
    return answer

arr = [1,2,3,4]
result = solution(arr)
print(result)

arr = [5,5]
result = solution(arr)
print(result)
````
